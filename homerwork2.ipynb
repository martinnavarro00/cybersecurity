## Exercise 1:

1. Suppose a password is chosen as a concatenation of seven lower-case dictionary words. Each word is selected uniformly at random from a dictionary of size 50,000. An example of such a password is "mothercathousefivenextcrossroom". How many bits of entropy does this have?
* In cyber security Entropy is given by the following equation: $$\ Entropy = log_2 (Number of posible calculations)$$
* In this case the Number of posible calculations is: $$\ 50,000^7$$ since there are seven words in the password.
* $$\ Entropy = 7* log_2(50,000) $$
* $$\ Entropy = 7* 15.60 $$
* $$\ Entropy = 109.26 bits $$

2. Consider an alternative scheme where a password is chosen as a sequence of 10 random alphanumeric characters (including both lower-case and upper-case letters). An example is "dA3mG67Rrs". How many bits of entropy does this have?
* In this case we have to consider that we have 26 lower case letter, 26 upper case letters and 10 numbers(from 0-9). 26+26+10=62
* $$\ Entropy = log_2(62^10) $$
* $$\ Entropy = 10 * log_2(62)$$
* $$\ Entropy = 59.53 bits $$

3. Which password is better, the one from 1. or 2.?
* In this case password number 1 is better since it has a higher entropy than password number 2, meaning it is more difficult to break.

1. Define what a Message Authentication Code (MAC) is and how it is used in cryptography.
* MAC is a technique used in cryptography that asures authenticity and integrity of a message between the sender and the receiver. The sender generates the MAC using a secret key and include the MAC in the message. The receiver also generate a MAC using the same secret key the sender used and compares it to verify the message has not been interceped by anyone else.
2. Explain the process of generating and verifying a MAC.
* To start the process both the sender and the receiver must share a confidential secret that is only shared between both of them. Then the sender use a hash function and the secret key to genereted the MAC and is send to the receiver.
* The receiver get the MAC, the message and the key. Then using the hash function the receiver uses the key to calculate the MAC from the message received.
* Finally the receiver compared the MAC he calculated and the MAC he received in order to determine its autenticity.
3. Discuss the importance of using MACs in secure communication systems.
* MACs are used to verifiy the authenticity and integrity of the message and the people who sends them. By having a matching MAC the receiver can verifty that the message is not alterate and can verify that the sender is indeed true.
